```python
tips = sns.load_dataset("tips")
tips
```

## first 5 lines of tips data

```python
tips.head(5)
```

## below you can see the larger the bill is the larger the tip is

```python
sns.relplot(x="total_bill",y="tip", hue="tip", data=tips)
```

## below you can see even the larger groups of people do not tip higher. 

```python
size_sums = tips.groupby("size", as_index=False).sum()
sns.relplot(x='size', y='tip', data=size_sums)
```

below you can see an assumtion that the higher th egrou pthe lower the tip.

```python
sns.lmplot(x='size', y='tip', data=size_sums)
```

## observations Viz-a:
1- the larger the bill the larger the tip
2- the larger the group the samller the tip

```python
car_crashes = sns.load_dataset("car_crashes")
car_crashes
```

## first 5 data list

```python
car_crashes.head(5)
```

## below you can see the faster you are speeding over the speed limit th emore car crashes

```python
sns.relplot(x="total",y="speeding", hue="speeding", data=car_crashes)
```

## below you can see that even drivers with no previous infractions crash more when speeding

```python
no_previous_sums = car_crashes.groupby("no_previous", as_index=False).sum()
sns.relplot(x='no_previous', y='total', data=no_previous_sums)
```

## below you can see the prediction that the faster you are speeding the more likely you will crash

```python
sns.lmplot(x='no_previous', y='total', data=no_previous_sums)
```

## observations Viz-b:
1- the faster drivers speed over the limit the more likely they will crash
2- even drivers with no previous infraction speed over the limit the more likely they will crash
